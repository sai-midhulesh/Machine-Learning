```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.mixture  import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
iris = datasets.load_iris()
print(iris)
```

```python
import pandas as pd
scalar = StandardScaler()
X = pd.DataFrame(iris.data)

x = scalar.fit_transform(X)
kmeans = KMeans(n_clusters = 3,random_state = 42)
kme = kmeans.fit_predict(x)
gm = GaussianMixture(n_components=3,random_state = 42)
gme = gm.fit_predict(x)

print(gme)
print(kme)

```

```python
    import pandas as pd
    import numpy as np
    import statsmodels.api as sm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,100)
y = np.sin(x) + np.random.normal(0,0.2,100)
low = sm.nonparametric.lowess(y,x)
x_ = low[:,0]
y_ = low[:,1]
plt.scatter(x,y,color='blue',label='original')
plt.plot(x_,y_,label='lowness',color='red')
plt.xlabel('x-axis')
plt.legend()
plt.show()
```

```python
import networkx as nx
import random
import matplotlib.pyplot as plt

g = nx.Graph()
for i in range(0,20):
    g.add_node(i)
count = 0
while count<=30:
    n1 = random.choice(list(g.nodes))
    n2 = random.choice(list(g.nodes))
    if (n1!=n2 and g.has_edge(n1,n2)!=True):
        g.add_edge(n1,n2)
        count+=1
nx.draw_networkx(g,with_labels=True)
plt.show()
    
```
